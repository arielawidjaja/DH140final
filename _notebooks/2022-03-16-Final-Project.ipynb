```python
import pandas as pd
```

```python
import video_games
video_game_df = pd.read_csv('video_games.csv')
```

Here is a summary of statistics for the video game dataset used for this project:

```python
video_game_df.describe()
```

The brief summary of statistics shows numerical data in relation to their corresponding categories. 
Several data to be noted are that the length of play time, either leisurely, rushed, or everything else, all have a minimum of 0. This can mean that the games were purchased but not played at all. In addition, the standard deviation for play time data vary, showing that the difference in play time differ depending on the play style (casual, rushed, story only, completionist, etc.) Although some may consider this data as outliers, they are valid data as it still shows the amount of time spent on the game, in which case is nonexistent (zero).

However, there are some notable outliers. All of the maximum play time in this data set is far above the average play time. Another notable outlier is the price for used games. The average price for all the used games is 17 dollars, yet the maximum reaches near 50. 

Here are some plots to show the significance of these outliers:

```python
import matplotlib.pyplot as plt
```

```python
#Plot for used price
video_game_df.sort_values(by='Metrics.Used Price',ascending=True).plot(kind='box',
             y='Metrics.Used Price')
```

```python
#Plot for average play time, all styles
video_game_df.sort_values(by='Length.All PlayStyles.Average',ascending=True).plot(kind='box',
             y='Length.All PlayStyles.Average')
```

```python
#Plot for average play time, main story only
video_game_df.sort_values(by='Length.Main Story.Average',ascending=True).plot(kind='box',
             y='Length.Main Story.Average')
```

```python
#Plot for average play time, main story and side quests
video_game_df.sort_values(by='Length.Main + Extras.Average',ascending=True).plot(kind='box',
             y='Length.Main + Extras.Average')
```

```python
video_game_df.sort_values(by='Metrics.Used Price',ascending=True).plot(kind='barh',
             x='Metadata.Genres',
             y='Metrics.Used Price',
             figsize=(15,150))
```

The graph is sorted this way to make it easier to see which genres are on the higher or lower in terms of price.

Based on the graph, the genres with higher prices are action, simulation, and adventure.

```python
video_game_df.sort_values(by='Length.All PlayStyles.Average',ascending=True).plot(kind='barh',
             x='Metadata.Genres',
             y='Length.All PlayStyles.Average',
             figsize=(15,150))
```

The graph is sorted this way to make it easier to see which genres are generally played longer.

Based on the graph, the genres with genrally longer playtimes are action, strategy, and simulation.

```python
video_game_df.sort_values(by='Length.Main Story.Average',ascending=True).plot(kind='barh',
             x='Metadata.Genres',
             y='Length.Main Story.Average',
             figsize=(15,150))
```

The graph is sorted this way to make it easier to see which genres are played longer if the players only played the main story.

Based on this graph, the genres with longer play time in regards of main story completion are strategy, simulation, and role-playing.

```python
video_game_df.sort_values(by='Length.Main + Extras.Average',ascending=True).plot(kind='barh',
             x='Metadata.Genres',
             y='Length.Main + Extras.Average',
             figsize=(15,150))
```

The graph is sorted this way to make it easier to see which genres are played longer if the players played the side quests in addition to the main story.

Based on this graph, the genres with longer play time for completion of the main story and extras are sports, simulation, and role-playing.

Overall, the simulation genre seems to be on top of the list of almost, if not all in terms of price and play time length. So far, it seems to be a popular genre that sells well.
